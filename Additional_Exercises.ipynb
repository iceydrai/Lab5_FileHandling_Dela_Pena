```python
import os

folder_path = r"C:\Users\ADRIAN\OneDrive\Desktop\Dela_Peña_Activty_5"

results = []

#  Loop through all .txt files
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r", encoding="utf-8") as f:
            lines = f.readlines()

        #  ignore empty lines
        lines = [line.strip() for line in lines if line.strip()]

        #  metrics
        num_lines = len(lines)
        words = sum(len(line.split()) for line in lines)
        characters = sum(len(line.replace(" ", "")) for line in lines)

        words_per_line = words / num_lines if num_lines else 0
        chars_per_word = characters / words if words else 0

        results.append([
            filename,
            num_lines,
            words,
            characters,
            words_per_line,
            chars_per_word
        ])

#  safety check
if not results:
    print("No .txt files found in the folder.")
else:
    # sort by word density (Words/Line)
    results.sort(key=lambda x: x[4], reverse=True)

    #  Print table header
    print(f"{'Filename':30} {'Lines':6} {'Words':6} {'Chars':8} {'W/Line':8} {'C/Word':8}")
    print("-" * 80)

    # Print rows
    for r in results:
        print(f"{r[0]:30} {r[1]:6} {r[2]:6} {r[3]:8} {r[4]:8.2f} {r[5]:8.2f}")

    # 🏆 Results
    print("\n🏆 Most content-dense file:", results[0][0])
    print("📉 Least content-dense file:", results[-1][0])

    #  Save CSV
    csv_path = os.path.join(folder_path, "content_summary.csv")

    with open(csv_path, "w", encoding="utf-8") as f:
        f.write("Filename,Lines,Words,Characters,WordsPerLine,CharsPerWord\n")
        for r in results:
            f.write(f"{r[0]},{r[1]},{r[2]},{r[3]},{r[4]:.2f},{r[5]:.2f}\n")

    print("\n💾 Saved to:", csv_path)
```

```python
import os
import string
from collections import Counter

# 📁 CHANGE THIS to your actual folder
folder_path = r"C:\Users\ADRIAN\OneDrive\Desktop\Dela_Peña_Activty_5"

# 📄 Load stopwords (optional)
stopwords_path = os.path.join(folder_path, "stopwords.txt")
stopwords = set()

if os.path.exists(stopwords_path):
    with open(stopwords_path, "r", encoding="utf-8") as f:
        stopwords = set(word.strip().lower() for word in f)

# 📊 store total frequency + per-file frequency
total_counter = Counter()
file_counters = {}

# 🔄 process all txt files
for filename in os.listdir(folder_path):
    if filename.endswith(".txt") and filename != "stopwords.txt":
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read().lower()

        # 🧹 remove punctuation
        text = text.translate(str.maketrans("", "", string.punctuation))

        words = text.split()

        # 🚫 remove stopwords
        filtered_words = [w for w in words if w not in stopwords]

        # 📊 count words
        counter = Counter(filtered_words)

        file_counters[filename] = counter
        total_counter.update(counter)

# 🚨 safety check
if not total_counter:
    print("No valid words found.")
else:
    # 🏆 TOP WORDS OVERALL
    print("\n🔥 Top 15 Most Frequent Words (All Files):")
    for word, count in total_counter.most_common(15):
        print(f"{word}: {count}")

    # 📂 OPTIONAL: per-file breakdown
    print("\n📁 Word Frequency Per File:")
    for filename, counter in file_counters.items():
        print(f"\n{filename}:")
        for word, count in counter.most_common(5):
            print(f"  {word}: {count}")

    # 💾 EXPORT CSV
    student_id = "2025-XXXX"  # ✏️ change this
    csv_path = os.path.join(folder_path, f"word_frequency_{student_id}.csv")

    with open(csv_path, "w", encoding="utf-8") as f:
        f.write("Word,Count\n")
        for word, count in total_counter.items():
            f.write(f"{word},{count}\n")

    print("\n💾 Saved to:", csv_path)
```

```python
import os
import shutil
from datetime import datetime

# 👤 YOUR INFO
student_id = "TUPM-25_0709"
student_name = "Adrian Elvyn L. Dela Peña"

# 📁 BASE FOLDER (your activity folder)
base_folder = r"C:\Users\ADRIAN\OneDrive\Desktop\Dela_Peña_Activty_5"

# 📄 AUTO-SELECT FIRST .txt FILE
txt_files = [f for f in os.listdir(base_folder) if f.endswith(".txt")]

if not txt_files:
    print("❌ No .txt files found in folder.")
else:
    # use first txt file automatically
    filename = txt_files[0]
    source_file = os.path.join(base_folder, filename)

    # 📁 CREATE BACKUP FOLDER
    backup_folder = os.path.join(base_folder, f"backup_{student_id}")
    os.makedirs(backup_folder, exist_ok=True)

    # ⏱️ TIMESTAMP
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # 📦 NEW BACKUP FILE NAME
    name, ext = os.path.splitext(filename)
    backup_filename = f"{name}_{student_id}_{timestamp}{ext}"
    backup_path = os.path.join(backup_folder, backup_filename)

    # 📌 COPY FILE
    shutil.copy2(source_file, backup_path)

    # 📊 FILE SIZE
    file_size = os.path.getsize(backup_path)

    # 📜 LOG FILE
    log_file = os.path.join(backup_folder, f"backup_log_{student_id}.txt")

    with open(log_file, "a", encoding="utf-8") as log:
        log.write("====================================\n")
        log.write(f"Student ID: {student_id}\n")
        log.write(f"Student Name: {student_name}\n")
        log.write(f"Original File: {filename}\n")
        log.write(f"Timestamp: {timestamp}\n")
        log.write(f"File Size: {file_size} bytes\n")
        log.write(f"Backup Path: {backup_path}\n")

    # ✅ OUTPUT
    print("✅ Backup completed successfully!")
    print(f"Student: {student_id} ({student_name})")
    print(f"File backed up: {filename}")
    print(f"Saved as: {backup_filename}")
    print(f"Log updated at: {log_file}")
```
